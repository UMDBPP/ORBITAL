```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
val1=7000
val2=25000
data=pd.read_csv('PTER04.csv')
gyy=np.abs(data['gyro y'])[val1:val2]
gyy=gyy*60/2/np.pi
alt=data['msAlt(M)'][val1:val2]

n=1000
#avy=moving_average(gyy)
avy=gyy.rolling(n).mean()
std=gyy.rolling(n).std()
fig,ax1=plt.subplots()
ax1.plot(gyy,ls='None',marker=".")
ax1.plot(avy)
ax1.plot(avy+std)
ax1.plot(avy-std)

ax2 = ax1.twinx()
ax2.plot(alt,color='orange')
print(np.max(avy))
print(np.max(avy+std))
print(np.max(gyy))
```
